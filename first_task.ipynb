```python
# Flatten the array to a 1D array
energy = ak.ravel(energy)
print (energy)
```

```python
""" fig, ax are the return variables. "plt.subplots" creates a new figure and a set of axes, allowing specification of various parameters to 
control the appearance and size of the figure and axes. "figsize" is the dimension of the plot in inches (width, height). "DPI" (dots per 
inch) for the figure. Higher DPI value results in a higher-resolution """
fig, ax = plt.subplots(figsize=(5, 3), dpi=200)
# energy is in MeV, we divide by 1e3 we want to plot in GeV
energy_to_plot = energy/1e3
# printing "energy_to_plot" gives a range of the values in the x-axis)
print(energy_to_plot)
# we define bins as a range from 0 to 0.1(range from "energy_to_plot" with 0.005(20 bins) bin width
bins = np.arange(0, 0.1, 0.005)
#steps should be less than stop. The jumps that it takes when plotting the data
print(f"Number of bins: {len(bins)}")
ax.hist(energy_to_plot, label='Genie example', bins=bins)

ax.set_xlabel('ECal Reconstructed Energy')
ax.set_ylabel('Density')
ax.set_yscale('log')
ax.legend()
plt.show()
```

```python
# second way of creating a histogram. This can be useful to find where majority of the values lies.
h = hist.Hist(
    # create a rec_energy axis, ranging from 0 to 6 with 60 bins
    hist.axis.Regular(60, 0, 6, name="rec_energy")
    
)
h.fill(rec_energy=energy/1e3)
```

```python
def get_data(filename, branches_to_read):
    ldmx_dict = {}
    file = uproot.open(f"{filename}:LDMX_Events")
    for key, branch in branches_to_read.items():
        ldmx_dict[key] = file[branch].array(library='ak')
    return ldmx_dict
    
```

```python
# https://uproot.readthedocs.io/en/latest/basic.html#reading-many-files-into-big-arrays
ldmx_dict = get_data(
    filename="/Users/deep/Downloads/egps_upstreamtagger_8_1_gev.root",
    branches_to_read = {"energy": 'EcalRecHits_electron/EcalRecHits_electron.energy_'},
)
ldmx_dict
```
